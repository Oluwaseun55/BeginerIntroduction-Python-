```python
import matplotlib.pyplot as plt

class Circle(object):
    def __init__(self, radius=3, color='red'):
        self.radius = radius
        self.color = color

    def add_radius(self, r):
        self.radius += r
        return self.radius

    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()

# Create an instance of Circle
RedCircle = Circle(10, 'blue')


# Print the initial radius
print('Initial radius of object:', RedCircle.radius)

# Draw the initial circle
RedCircle.drawCircle()

# Add radius and print the updated radius
RedCircle.add_radius(2)
print('Radius of object after applying the method add_radius(2):', RedCircle.radius)

# Draw the circle after adding radius
RedCircle.drawCircle()

# Add more radius and print the updated radius
RedCircle.add_radius(5)
print('Radius of object after applying the method add_radius(5):', RedCircle.radius)

# Draw the circle after adding more radius
RedCircle.drawCircle()
```

## The Rectangle Class

Let's create a class rectangle with the attributes of height, width, and color. We will only add the method to draw the rectangle object:

```python
import matplotlib.pyplot as plt

class Rectangle(object): 
    
    def __init__(self, width, height, color='teal'):
        self.width= width
        self.height= height
        self.color= color
        
    def drawRectangular(self):
        plt.gca().add_patch(plt.Rectangle((0,0), self.width, self.height, fc=self.color))
        plt.axis('scaled')
        plt.show()

SkinnyBlueRectangular = Rectangle(4, 2,'teal')

SkinnyBlueRectangular.height

SkinnyBlueRectangular.width

SkinnyBlueRectangular.color

SkinnyBlueRectangular.drawRectangular()
```

```python
SkinnyYellowRectangular = Rectangle(5, 2, 'yellow')

SkinnyYellowRectangular.drawRectangular()
```
